```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
```

# Load Metadata

```python
metadata_df = pd.read_csv('./data/loaded/metadata.csv', index_col=0)
#metadata_df['pose_name_l3'] = metadata_df.apply(lambda x: 'Intense Side Stretch Pose' if x['pose_name_alt_l3'] == 'Parsvottanasana' else x['pose_name_l3'], axis=1)
metadata_df.head()
```

```python
mean_px = pd.melt(metadata_df[['channel_B_pixel_mean', 'channel_G_pixel_mean', 'channel_R_pixel_mean']].rename(columns={'channel_B_pixel_mean': 'B', 'channel_G_pixel_mean': 'G', 'channel_R_pixel_mean': 'R'}), var_name='Channel', value_name='Mean Pixel Value')

channels = ['B', 'G', 'R']
colors = ['blue', 'green', 'red']

fig = go.Figure()

for channel, col in zip(channels, colors):
    fig.add_trace(go.Violin(y=mean_px[mean_px['Channel'] == channel]['Mean Pixel Value'], x=mean_px[mean_px['Channel'] == channel]['Channel'], box_visible=True, meanline_visible=True, name=channel, marker_color=col))
fig.update_layout(title='Distribution of Mean Pixel Value Across Image Dimension')
fig.write_html('./plots/mean_px.html')
```

```python
std_px = pd.melt(metadata_df[['channel_B_pixel_std', 'channel_G_pixel_std', 'channel_R_pixel_std']].rename(columns={'channel_B_pixel_std': 'B', 'channel_G_pixel_std': 'G', 'channel_R_pixel_std': 'R'}), var_name='Channel', value_name='Std. Dev of Pixel Value')

channels = ['B', 'G', 'R']
colors = ['blue', 'green', 'red']

fig = go.Figure()

for channel, col in zip(channels, colors):
    fig.add_trace(go.Violin(y=std_px[std_px['Channel'] == channel]['Std. Dev of Pixel Value'], x=std_px[std_px['Channel'] == channel]['Channel'], box_visible=True, meanline_visible=True, name=channel, marker_color=col))
fig.update_layout(title='Distribution of Std. Deviation of Pixel Values Across Image Dimension')
fig.write_html('./plots/std_px.html')
```

```python
min_px = pd.melt(metadata_df[['channel_B_pixel_min', 'channel_G_pixel_min', 'channel_R_pixel_min']].rename(columns={'channel_B_pixel_min': 'B', 'channel_G_pixel_min': 'G', 'channel_R_pixel_min': 'R'}), var_name='Channel', value_name='Min Pixel Value')

channels = ['B', 'G', 'R']
colors = ['blue', 'green', 'red']

fig = go.Figure()

for channel, col in zip(channels, colors):
    fig.add_trace(go.Violin(y=min_px[min_px['Channel'] == channel]['Min Pixel Value'], x=min_px[min_px['Channel'] == channel]['Channel'], box_visible=True, meanline_visible=True, name=channel, marker_color=col))
fig.update_layout(title='Distribution of Min Pixel Values Across Image Dimension')
fig.write_html('./plots/min_px.html')
```

```python
max_px = pd.melt(metadata_df[['channel_B_pixel_max', 'channel_G_pixel_max', 'channel_R_pixel_max']].rename(columns={'channel_B_pixel_max': 'B', 'channel_G_pixel_max': 'G', 'channel_R_pixel_max': 'R'}), var_name='Channel', value_name='Max Pixel Value')

channels = ['B', 'G', 'R']
colors = ['blue', 'green', 'red']

fig = go.Figure()

for channel, col in zip(channels, colors):
    fig.add_trace(go.Violin(y=max_px[max_px['Channel'] == channel]['Max Pixel Value'], x=max_px[max_px['Channel'] == channel]['Channel'], box_visible=True, meanline_visible=True, name=channel, marker_color=col))
fig.update_layout(title='Distribution of Max Pixel Values Across Image Dimension')
fig.write_html('./plots/max_px.html')
```

```python
pixel_stats = pd.concat([mean_px, std_px, min_px, max_px], axis=1).iloc[:, [0, 1, 3, 5, 7]]
pixel_stats = pd.melt(pixel_stats.rename(columns={'Mean Pixel Value': 'Mean', 'Std. Dev of Pixel Value': 'Std. Dev', 'Min Pixel Value': 'Min', 'Max Pixel Value': 'Max'}), id_vars='Channel', var_name='Statistic', value_name='Pixel Value')
fig = px.box(pixel_stats, y='Pixel Value', x='Statistic', color='Channel', color_discrete_sequence=['blue', 'green', 'red'])
fig.update_layout(title='Distribution of Pixel Values Across Image Channel Over Different Statistics')
fig.write_html('./plots/all_stat_px.html')
```

```python
top_classes = metadata_df['pose_name_l3'].value_counts().index[:40]
fig = px.bar(metadata_df['pose_name_l3'].dropna().map(lambda x: x if x in top_classes else 'Others').value_counts(), orientation='h', log_x=True)
fig.update_xaxes(title='Count (Log Scale)')
fig.update_yaxes(title='Pose Name')
fig.update_layout(title='Distribution of Poses', showlegend=False)
fig.write_html('./plots/pose_dist.html')
```

```python
metadata_df['height_px'] = metadata_df['image_shape'].map(lambda x: eval(x)[0] if type(x) != float else x)
metadata_df['width_px'] = metadata_df['image_shape'].map(lambda x: eval(x)[1] if type(x) != float else x)
fig = px.scatter(metadata_df[['height_px', 'width_px', 'pose_name_l3']].dropna().rename(columns={'height_px': 'Height (Pixels)', 'width_px': 'Width (Pixels)', 'pose_name_l3': 'Pose Name'}), y='Height (Pixels)', x='Width (Pixels)', color='Pose Name')
fig.update_layout(title='Distribution of Image Size Across Different Poses')
fig.write_html('./plots/img_size_dist.html')
```
